```python
# Define Reference Frames
N = ReferenceFrame('N')
F = ReferenceFrame('F')
P = ReferenceFrame('P')
```

```python
# Define Symbolic Variables
x, theta, delta = dynamicsymbols('x theta delta')
l_p,d,R = symbols('l_p,d,R')
```

```python
# Define Positions
No = Point('No')
I = Point('I')
I.set_pos(No, x*F.x)
Po = Point('Po')
Po.set_pos(I, l_p*P.z)
LN = Point('LN')
LN.set_pos(I, d*P.y - R*F.y)
RN = Point('RN')
RN.set_pos(I, -d*P.y - R*F.y)
```
